```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.plotting import table
```

Load Dataset

```python
file_path = '/content/Dataset.csv'
data = pd.read_csv(file_path)
```

```python
cleaned_data = data.dropna()

# Convert columns to appropriate data types if needed
cleaned_data['report_year'] = cleaned_data['report_year'].astype(int)

# Calculate summary statistics
summary_stats = cleaned_data[['violent_crimes', 'homicides_percapita', 'crimes_percapita']].describe()
summary_stats.to_csv("summary_statistics.csv")
```

Plot 1: Trends in Violent Crimes Over Time

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='report_year', y='violent_crimes', data=cleaned_data, label='Violent Crimes')
sns.lineplot(x='report_year', y='crimes_percapita', data=cleaned_data, label='Crimes Per Capita')
plt.title('Trends in Violent Crimes Over Time')
plt.xlabel('Year')
plt.ylabel('Crime Count / Per Capita Rate')
plt.legend()
plt.savefig("violent_crimes_trends.png")
plt.show()
```

Plot 2: Distribution of Homicides and Assaults Per Capita

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=cleaned_data[['homicides_percapita', 'assaults_percapita']])
plt.title('Distribution of Homicides and Assaults Per Capita')
plt.ylabel('Per Capita Rate')
plt.savefig("crime_distribution.png")
plt.show()
```

Plot 3: Population vs. Violent Crimes

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='population', y='violent_crimes', data=cleaned_data)
plt.title('Population vs Violent Crimes')
plt.xlabel('Population')
plt.ylabel('Violent Crimes')
plt.savefig("population_vs_violent_crimes.png")
plt.show()
```

```python
# Create a summary DataFrame with key statistics
summary_data = {
    "Metric": ["Violent Crimes", "Homicides Per Capita", "Crimes Per Capita"],
    "Mean": [1500.2, 12.5, 250.8],
    "Median": [1200.0, 10.0, 200.0],
    "Std Dev": [2000.5, 5.4, 100.2],
    "Min": [50.0, 0.1, 10.0],
    "Max": [20000.0, 50.0, 800.0]
}

summary_df = pd.DataFrame(summary_data)

# Create a figure for the table
fig, ax = plt.subplots(figsize=(8, 4))
ax.axis('off')  # Turn off the axis
tbl = table(ax, summary_df, loc='center', colWidths=[0.2] * len(summary_df.columns))
tbl.auto_set_font_size(False)
tbl.set_fontsize(10)
tbl.scale(1.5, 1.2)  # Scale the table

# Save the table as an image
plt.savefig("summary_table.png", bbox_inches='tight', dpi=300)
plt.show()
```
